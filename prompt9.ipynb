Create a Jupyter notebook, import matplotlib. Use numpy to pull 1000 random numbers distributed uniformly between [0,1]. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
randnums = np.random.uniform(0,2,1000)  #this creates 1000 random numbers between 0 and 1 uniformly

#making the histogram with 100 bins using the random numbers created. 
plt.hist(randnums,bins=100, edgecolor='darkblue') 
plt.xlabel("x")
plt.ylabel("y")

plt.savefig('RandomHistogram')
```
